```python
import numpy as np
import pandas as pd
from numpy import random
import matplotlib.pyplot as plt
```

```python
x1 =  np.random.randint(low=1,high=20,size=(50,))
x1
                     
                     

                     
```

```python
x2 =  np.random.randint(low=1,high=20,size=(50,))
x2
                     
     
```

```python
y=3+5*x1-4*x2+np.random.random(50)
y
```

```python
x = pd.DataFrame({'intercept':np.ones(50),'x1':x1,'x2':x2})
x
```

```python
x.shape
```

```python
x.sum()
```

```python
x.describe()
```

```python
w=np.random.random(x.shape[1])
w
```

```python
def myprediction(feature,weight):
    prediction=np.dot(feature,weight)
    return prediction
myprediction(x,w)
```

```python
def error(target,feature,weight):
    error_value=target-myprediction(feature,weight)
    return error_value
error(y,x,w)
```

```python
def cost(target,feature,weight):
    error_value=error(target,feature,weight)
    transpose=error_value.T
    cost_value=np.dot(transpose,error_value)
    return cost_value
cost(y,x,w)
    
```

```python
def gradient(target,feature,weight):
    e=error(target,feature,weight)
    grand_value=-2*np.dot(feature.T,e)
    return grand_value
gradient(y,x,w)
```

```python
def linear_regression(target,features,learning_rate,num_steps):
    weight=np.random.random(x.shape[1])
    for i in range(num_steps):
        weight=weight-learning_rate*gradient(target,features,weight)
        if(i%1000==0):
            print(cost(target,features,weight),weight)
        
        
        
    return weight

linear_regression (y,x,0.00001,100000)
```
