```python
def create_plot(ptype):
    x=np.arange(0,10,0,1)
    
    if ptype == "linear":
        y=x
    elif ptype== "quadratic":
        y=x**2
    elif ptype==" cubic":
        y=x**3
    plt.plot(x,y)
    
create_plot("linear") 
create_plot("quadratic")
create_plot("cubic")
    
    
    
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
font1={"family":"serif","color":"blue","size":10}
plt.plot(x,y, alpha=0.2)
plt.plot(x,y,linestyle="dashed",color="red",linewidth=2,marker="o",markerfacecolor="green")
plt.axhline(y=5,color="black",linestyle = '-')
plt.axvline(x=5,color="black",linestyle = '-')
plt.xlabel("input",fontdict=font1)
plt.ylabel("output")
plt.title("line chart",loc="left")
plt.show()
```

```python
# x-axis values
x = [1, 2, 3, 4, 5]

# Y-axis values
y = [2, 4, 6, 8, 18]

font1 = {'family':'serif','color':'blue','size':10}


# Function to plot
plt.plot(x,y,alpha=0.2) # use alpha to change the opacity of the line

plt.plot(x,y,linestyle = 'dashed',color='red',linewidth=2,marker = 'o',markerfacecolor = 'green', markersize = 10)

#plt.plot(y,x,ls = '--',c='b',lw=10)

# drawing multiple lines
#plt.plot(x, y, y, x)

#plot a horizontal line
plt.axhline(y = 5, color = 'black', linestyle = '-')

plt.axvline(x= 5, color = 'black', linestyle = '-')


plt.xlabel("input",fontdict=font1)     # change the font of title

plt.ylabel("output")
plt.title("Line Chart",loc = 'left')   # default location is center

#plt.grid(color = 'green', linestyle = '--', linewidth = 1)

# function to show the plot
plt.show()
```

```python
# bar graph
x1 = ['A', 'B', 'C', 'D']
y1 = np.array([10, 20, 10, 30])
y2 = np.array([20, 25, 15, 25])
y3 = np.array([12, 15, 19, 6])
y4 = np.array([10, 29, 13, 19])
plt.bar(x1, y1, color='r')

plt.bar(x1, y2, bottom=y1, color='b')

plt.bar(x1, y3, bottom=y1+y2, color='y')

plt.bar(x1, y4, bottom=y1+y2+y3, color='g')

plt.xlabel('x-axis', fontweight ='bold', fontsize = 15)


plt.show()
# plot bars in stack manner
# horizontal bars
plt.barh(x1, y1,color="red")
# back to back plots
plt.barh(x1, -y1, color = 'b')

plt.show()
```
